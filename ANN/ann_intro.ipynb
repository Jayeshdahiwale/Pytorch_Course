##### Multi Layer Perceptron Model 

To build the network of perceptrons , we can connect the layers of perceptrons, using Multi Layer Perceptron Model.
The output of one perceptron is fed into input of another perceptron.

The first layer is called input layer. The input data can be in tabular form
The last layes is called output layer. The last layer can have more than one neuron depending on type of expected output.
Layers between input and output layer are called hidden layers.

Neural networks become deep neural netwrork when it 2 or more hidden layers


##### Activation Functions

Some really common activation functions

1. The most simple networks rely on basic step size functions.  If x > 0 then f(x) = 1 ,else f(x) =0
$$
f(x) =
\begin{cases}
  1 & \text{if } x > 0 \\
  0 & \text{otherwise}
\end{cases}
$$

2. Sigmoid function. It also gives the values in the range of 0 to 1. It is a continuous function.
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

3. Hyperbolic Tangent function : It gives the values in the range of -1 to 1. 
$$ \sinh x = \frac{e^x - e^{-x}}{2} $$
$$ \cosh x = \frac{e^x + e^{-x}}{2} $$
$$ \tanh x = \frac{\sinh x}{\cosh x} = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

4. Rectified linear unit  : It is basically a very simple activation function. Eg . max(0,x)
It is used in Convolutional layers.
$$
\text{ReLU}(x) = \max(0, x)
$$
Relu is found to have very good result especially when we are dealing with vanishing gradient problem.


##### Multi Class Classifications

There are two types of multiclass classifications.
Non-exclusive classes : A data point can have multiple classes/cateogires assigned to it.
Example : Photos can have multiple tags( beach ,family, vacation etc.)
Mutually Exclusive Classes : Only one class per data point.
Example : Photos can be assigned as black and white or full colour.


In case of Non-exclusinve classes, we have to use multiple neurons in the output layer. And the activation function we should use is continuous like sigmoind function.
In case of mutually exclusive classes, we should use softmax function. Bascially it calcualates the probability of happening of particualt output.

$$
\text{softmax}(\mathbf{x})_i = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}
$$



##### Cost Function And Gradient Descent

Most often used Cost function is C = 1/2n * (y(x) - al(x))^2, where y(x) is true value. And al(x) if output layer l, neurons output. 
$$
C = \frac{1}{2n} \sum_{i=1}^{n} (y(x_i) - a_l(x_i))^2
$$

squaring the error does 2 useful things, first it punishes the error hugely and keeps the difference always positive.

Now as we know a neural networks is the combination of lot of weights. So how do we get to know what value of  weight will lead us  to minimize the cost function.
Answer is Gradient Descent. 

Our Cost function will be very complex.


Another thing is cost function will be n dimensional. C(W1,W2,W3...,Wn). So I would use lot of computational power. But still we can find the wieght using gradient descent without using lot of computational power.


Steps : 1. We start with a point. Calculate slope on that point.


2. We move in the downward direction of the slope.


3. And we keep repeating this process until we converge to zero.


4. We can change our step size to find the next point.


5. Smaller step size take longer time to converge to minimum. This step size is knows as learning rate.


6. Larger step size process faster faster and takes less time but there is a risk of overshooting.


7. But we can be clever and adapt our step size as we go along. So we can go with larger steps, then go with smaller as we realizes that slope is getting closer to zero. This is known as Adaptive Gradient Descent.


8. In 2015, Kingma and Ba published their paper "Adam: A method for Schochastic Optimization"


9. When we are dealing with such n dimensional vectors the notation changes from derviative to gradient.


10. This mean we calculate 
 $$
\nabla C(w_1, w_2, \ldots, w_n)
$$

11. For classification problems we use cross entropy loss function.
$$
C(y, a) = - [y \log(a) + (1 - y) \log(1 - a)]
$$

12. For m number of classes M > 2 , 
$$
C(\mathbf{y}, \mathbf{a}) = - \sum_{i=1}^{m} y_i \log(a_i)
$$
